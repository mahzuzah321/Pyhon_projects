```python
import pandas as pd
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']], df.Outcome, train_size=0.8, random_state=2)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
model.score(X_test,y_test)
```
